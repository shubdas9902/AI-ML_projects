```python
parkinsondf = df.drop(columns=['name'])
#
```

```python
parkinsondf.groupby('status').mean()
#
```

```python
X=parkinsondf.drop(columns=['status'],axis=1)
Y=parkinsondf['status']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
scaler=StandardScaler()
scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
models=[LogisticRegression(),svm.SVC(kernel='linear')]
for i in range(len(models)):
    models[i].fit(X_train, Y_train)
    print(f'{models[i]} : {models[i].score(X_test, Y_test)}')
```
